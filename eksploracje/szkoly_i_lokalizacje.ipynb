```python
import pandas as pd
```

```python
szkoly = pd.read_csv("../dane/szkoly2014.csv", index_col="id_szkoly")
```

```python
szkoly.head(2)
```

```python
szkoly.columns
```

```python
szkoly = szkoly.rename(columns={col: col.replace("_szkoly", "") for col in szkoly})
```

```python
lokalizacje = pd.read_csv("../dane/lokalizacje_ponagdimnazjalnych_2014.csv", index_col="id_szkoly")
```

```python
lokalizacje.head()
```

```python
# licea w Wawie 
warszawskie = szkoly.query("miejscowosc == 'Warszawa' & typ == 'LO'")
```

```python
warszawskie = warszawskie[["nazwa", "adres", "miejscowosc", "pna", "gmina"]]
warszawskie = warszawskie.rename(columns={"pna": "kod_pocztowy", "gmina": "dzielnica"})
```

```python
warszawskie.to_csv("../dane/przetworzone/licea_warszawskie.csv")
```

```python
szkoly_z_lokalizacjami = szkoly.join(lokalizacje, how='right')
```

```python
szkoly_z_lokalizacjami.to_csv("../dane/przetworzone/ponadgim_z_lokalizacjami.csv")
```

```python
wojewodztwa = szkoly_z_lokalizacjami.groupby("wojewodztwo")[["lat", "lng"]].mean()
wojewodztwa['liczba_szkol'] = szkoly_z_lokalizacjami.groupby("wojewodztwo").size()
wojewodztwa.to_csv("../dane/przetworzone/wojewodztwa_lokalizcje.csv")
```

```python
gminy = szkoly_z_lokalizacjami.groupby(["wojewodztwo", "powiat", "gmina"])[["lat", "lng"]].mean()
gminy['liczba_szkol'] = szkoly_z_lokalizacjami.groupby(["wojewodztwo", "powiat", "gmina"]).size()
gminy.to_csv("../dane/przetworzone/gminy_lokalizcje.csv")
```
